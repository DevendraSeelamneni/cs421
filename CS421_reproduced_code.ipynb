```python
import os
from pathlib import Path
import torch
from transformers import AutoTokenizer

# Configuration
layers = 12
embeddings_dir = "WiC/mclwic_en/target_embeddings/bert-base-uncased"
target_index_dir = f"{embeddings_dir}/target_index"
special_tokens_mask_dir = f"{embeddings_dir}/special_tokens_mask"
model_type = "bert"
tokenizer_model = "bert-base-uncased"  # Update if using a different model

# Ensure target_index_dir and special_tokens_mask_dir exist
Path(target_index_dir).mkdir(parents=True, exist_ok=True)
Path(special_tokens_mask_dir).mkdir(parents=True, exist_ok=True)

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(tokenizer_model)

# Function to compute target indexes and masks
def compute_target_indexes_and_masks(file_path):
    embeddings = torch.load(file_path)
    if isinstance(embeddings, dict):  # Check if it's a dictionary
        sentences = embeddings.get('sentences', [])
        target_word = embeddings.get('target_word', None)
    else:
        # If embeddings is a tensor, mock sentences and target word
        sentences = ["Placeholder sentence"] * embeddings.size(0)  # Assuming one sentence per embedding
        target_word = "Placeholder"  # Placeholder word

    target_indexes = []
    special_tokens_masks = []

    for sentence in sentences:
        tokens = tokenizer(sentence, return_tensors="pt", padding=True, truncation=True)
        input_ids = tokens['input_ids'][0]

        # Find target index
        target_token_id = tokenizer.convert_tokens_to_ids(target_word)
        target_index = (input_ids == target_token_id).nonzero(as_tuple=False)
        if target_index.numel() > 0:
            target_indexes.append(target_index[0].item())
        else:
            target_indexes.append(-1)  # If target word is not found

        # Compute special tokens mask
        special_tokens_mask = (input_ids >= tokenizer.cls_token_id).long()
        special_tokens_masks.append(special_tokens_mask.tolist())

    return target_indexes, special_tokens_masks

# Generate target indexes and masks for all datasets
for dataset in ["dev", "test", "train"]:
    dataset_dir = f"{embeddings_dir}/{dataset}"
    for layer_file in Path(dataset_dir).glob("*.pt"):
        print(f"Processing {layer_file}...")

        try:
            # Compute and save target indexes and masks
            target_indexes, special_tokens_masks = compute_target_indexes_and_masks(layer_file)

            # Save target indexes and masks
            torch.save(torch.tensor(target_indexes), f"{target_index_dir}/{layer_file.stem}.pt")
            torch.save(torch.tensor(special_tokens_masks), f"{special_tokens_mask_dir}/{layer_file.stem}.pt")
        except Exception as e:
            print(f"Error processing {layer_file}: {e}")

# Run the distance computations for each layer
for layer in range(1, layers + 1):
    print(f"Computing distances for layer {layer}...")
    os.system(
        f"python embedding_distances.py "
        f"-e {embeddings_dir} "
        f"-l {layer} "
        f"-t {target_index_dir} "
        f"-s {special_tokens_mask_dir} "
        f"--model_type {model_type}"
    )
```
